# Data Cleaning

## حذف التكرار من الداتا

```python
# اشوف القيم المتكررة بشكل كامل (الصف المتكرر كامل)

df[df.duplicated()].sort_values(by = ['Job Title','Company Name','Salary Estimate'], ascending = True)

df.head()
```

```python
# حذف الصفوف المكررة ، ما عدا اول صف (الصف الاساسي الي تكرر البيانات تعته)

df = df.drop_duplicates(keep='first')
```

```python
print('عدد الصفوف بعد حذف الصفوف المتكررة' + ' - ' + str(len(df)))
```

## عمود الراتب

```python
df['Salary Estimate'].unique()
```

```python
# ازالة النص الزائد على رينج الراتب
df['Salary Range'] = df['Salary Estimate'].str.replace(r'\(Glassdoor est.\)|\(Employer est.\)', '', regex = True).str.strip()
```

```python
df['Salary Range'].unique()
```

## عمود التقييم

```python
# استبدال القيم -1 ب 0 ، مفش تقييم ل الشركة او تقييمها صفر
df['Rating'] = df['Rating'].replace(-1.0, 0)
```

```python
# عدد مرات تكرار قيمة 0
len(df[df['Rating'] == 0])
```

## عمود اسم الشركة

```python
# تنظيف اسم الشركة و ازالة اي رموز زائدة في الاسم
df['Company Name'] = df['Company Name'].str.replace(r'\n\d+(\.\d+)?', '', regex=True)
```

```python
df['Company Name'].unique()
```

## عمود الموقع

```python
df['Location'].unique()
```

```python
# معرفة شكل الداتا في عمود المكان
df['Location'][0].split(',')
```

```python
# فصل المدينة عن الولاية باستخدام عمود المكان و تخزين القيم ف يعمود جديد
def location_parsing(location):
    location = location.split(',')
    if len(location) >= 2:
        City = location[0].strip()
        State = location[1].strip()
        return City, State
    else:
        return None, None

df[['City', 'State']] = df['Location'].apply(location_parsing).apply(pd.Series)
```

```python
df[(df['City'].isnull()) & (df['State'].isnull())]['Location'].value_counts()
```

```python
# استبدال القيم الفارغة في المدينة و الولاية
df.loc[df['Location'].str.contains('Remote', case=False, na=False), ['City', 'State']] = 'Remote'

df.loc[df['Location'].str.contains('United States', case = False, na = False),['City','State']] = ['United States','US']

df.loc[df['Location'].str.contains('Utah', case = False, na = False),['City','State']] = ['Utah','UT']

df.loc[df['Location'].str.contains('New Jersey', case = False, na = False),['City','State']] = ['New Jersey','NJ']

df.loc[df['Location'].str.contains('Texas', case = False, na = False),['City','State']] = ['Texas','TX']

df.loc[df['Location'].str.contains('California', case = False, na = False),['City','State']] = ['California','CA']

df.loc[df['Location'].str.contains('Patuxent, Anne Arundel, MD', case = False, na = False),['City','State']] = ['Anne Arundel','MD']
```

## عمود المقر الرئيسي

```python
df['Headquarters'].str[-2:].unique()
```

```python
# استبدال القيم -1 ب قيمة المكان
def replace_headquarters(row):
    if row['Headquarters'] == '-1':
        return row['Location']
    else:
        return row['Headquarters']

df['Headquarters'] = df.apply(replace_headquarters, axis=1)
```

## سعة الموظفين

```python
# استبدال قيم -1 بقيمة مجهول
df['Size'] = df['Size'].replace('-1','Unknown')
```

## سنة التاسيس

```python
# استبدال قيم -1 بقيمة مجهول
df['Founded'] = df['Founded'].replace(-1,'Unknown')
```

## نوع ملكية الشركة

```python
# استبدال قيم -1 بقيمة مجهول
df['Type of ownership'] = df['Type of ownership'].replace('-1','Unknown')
```

## الصناعة

```python
# استبدال قيم -1 بقيمة مجهول
df['Industry'] = df['Industry'].replace('-1','Unknown')
```

## القطاع

```python
# استبدال قيم -1 بقيمة مجهول
df['Sector'] = df['Sector'].replace('-1','Unknown')
```

## الربح / العائد

```python
# استبدال قيم -1 بقيمة مجهول
df['Revenue'] = df['Revenue'].replace('-1','Unknown / Non-Applicable')
```

## المنافسين

```python
# حذف عمود المنافسين بسبب عدم اهميته
df = df.drop(columns=['Competitors'])
```

# Data Processing

## الراتب

```python
# انشاء عمود جديد يحتوى على القيمة الادنى والاعلى للراتب
df[['Min_Salary','Max_Salary']] = df['Salary Range'].str.split('-',expand = True)
```

```python
df['Min_Salary'].head()
```

```python
# ازالة الرموز و الاحرف من الرقم
df['Min_Salary'] = df['Min_Salary'].str.replace(r'\$', '', regex=True).str.replace(r'K', '', regex=True)
```

```python
df['Min_Salary'].head()
```

```python
# تغيير نوع البيانات لسلسلة الحد الأدنى من الراتب من كائن إلى عدد صحيح
df['Min_Salary'] = df['Min_Salary'].astype(int)

# ضرب القيم في 1000 لتحويلها إلى آلاف
df['Min_Salary'] = df['Min_Salary'] * 1000

# عرض الأسطر الأولى للتأكد من التغييرات
df['Min_Salary'].head()
```

```python
df['Max_Salary'].head()
```

```python
# ازالة الرموز و الاحرف من الرقم
df['Max_Salary'] = df['Max_Salary'].str.replace(r'\$', '', regex=True).str.replace(r'K', '', regex=True)
```

```python
# تغيير نوع البيانات لسلسلة الحد الاعلى من الراتب من كائن إلى عدد صحيح
df['Max_Salary'] = df['Max_Salary'].astype(int)

# ضرب القيم في 1000 لتحويلها إلى آلاف
df['Max_Salary'] = df['Max_Salary'] * 1000

# عرض الأسطر الأولى للتأكد من التغييرات
df['Max_Salary'].head()
```

```python
# انشاء عمود متوسط الراتب

df['Average_Salary'] = (df['Min_Salary'] + df['Max_Salary'])/2
```

```python
df['Average_Salary'] = df['Average_Salary'].astype(int)

df['Average_Salary'] = df['Average_Salary'] * 1000

df['Average_Salary'].head()
```

## خدمة الشركة في سوق العمل

```python
from datetime import date

def year_diff(founded_year):
    if founded_year == 'Unknown':
        return 'Unknown'
    else:
        return date.today().year - founded_year

df['Company age'] = df['Founded'].apply(year_diff)
```

```python
df['Company age'].head()
```

## انشاء عمود ل فئة الوظيفة

```python
# Mapping Job Titles with the relevant Job Categories.

category_mapping = {
    'Business Intelligence Analyst': 'BI Analyst',
    'Analyst': 'Data Analyst',
    'Scientist': 'Data Scientist',
    'Machine Learning': 'ML Engineer',
    'Data Engineer': 'Data Engineer'
}

# Creating a custom function to create a Job Category series.

def categorize_job_title(job_title):
    for keyword, category in category_mapping.items():
        if keyword.lower() in job_title.lower():
            return category
    return 'Other'

df['Job Category'] = df['Job Title'].apply(categorize_job_title)
```

```python
# القيم المتطرفة و معالجتها

data = df.loc[df['Job Category'] == 'Data Scientist']['Average_Salary']

q1, q3 = np.percentile(data, [25, 75])
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Find outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]
outliers = set(outliers)

print("Outliers:", outliers)
```

```python
# تنظيف البيانات من القيم المتطرفة
data = df.loc[df['Job Category'] == 'Data Scientist']['Average_Salary']


q1, q3 = np.percentile(data, [25, 75])

iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# تنظيف البيانات
cleaned_data = data[(data >= lower_bound) & (data <= upper_bound)]
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(cleaned_data, vert=True, labels=['Cleaned Data'])

plt.xlabel('Values')
plt.title('Box Plot After Removing Outliers')
plt.grid(True)
plt.show()
```

## تحليل المهارات الاكثر طلبا في الوظائف من خلال وصف الوظيفة

```python
# فحص مهارة sql
def contains_sql(job_desc):
    if 'sql'.lower() in job_desc.lower():
        return 1
    else:
        return 0

df['SQL'] = df['Job Description'].apply(contains_sql)
```

```python
# python
def contains_python(job_desc):
    if 'python'.lower() in job_desc.lower():
        return 1
    else:
        return 0

df['Python'] = df['Job Description'].apply(contains_python)
```

```python
# Data Visualization
def contains_viz(job_desc):
    if 'visualization'.lower() in job_desc.lower():
        return 1
    else:
        return 0

df['Data Visualization'] = df['Job Description'].apply(contains_viz)
```

```python
# Statistics
def contains_stats(job_desc):
    if 'statistics'.lower() in job_desc.lower():
        return 1
    else:
        return 0

df['Statistics'] = df['Job Description'].apply(contains_stats)
```

```python
# Machine Learning
def contains_ml(job_desc):
    if 'machine learning'.lower() in job_desc.lower():
        return 1
    else:
        return 0

df['Machine Learning'] = df['Job Description'].apply(contains_ml)
```
